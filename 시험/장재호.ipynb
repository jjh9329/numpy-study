```python
import pandas as pd
import numpy as np
#- 주진 전체 기간의 각 나라별 골 득점수 상위 5개 국가와 그 득점수를 출력하시오
df = pd.read_csv('worldcupgoals.csv')
df
```

주진 전체 기간의 각 나라별 골 득점수 상위 5개 국가와 그 득점수를 출력하시오
    

```python
df['Country'].unique()
df.groupby('Country')['Goals'].sum().sort_values(ascending=False).head()
```

주어진 전체기간동안 골득점을 한 선수가 가장 많은 나라 상위 5개 국가와 그 선수 숫자를 출력하시오.

```python
df.groupby('Country')['Player'].count().sort_values(ascending=False).head()
```

    - Years 컬럼은 년도 -년도 형식으로 구성 되어있고, 각 년도는 4자리 숫자이다. 각 년도 표기가 4자리 숫자로 안된 케
    이스가 존재한다. 해당 건은 몇 건인지 출력하시오.


```python
df[df['Years'].str.replace('-','').astype('int64') <1000]
```

```python
# - 3번 문제에서 발생한 예외 케이스(년도 표기가 4자리 숫자가 아닌 경우)를 제외한 새로운 데이터 프레임을 생성하고
#    전체 데이터 행의 숫자를 출력하시오.
```

```python
df2 = df[df['Years'].str.replace('-','').astype('int64') >1000]
df2
```


    - 월드컵 출전횟수를 나타내는 ‘LenCup’ 컬럼을 추가하고 4회 출전한 선수의 숫자를 구하시오.


```python
a =df2['Years'].str.replace('-','').astype('int64')
#df2[np.logical_and(a>10000 ,a<100000000)] = 1
df2['LenCup'] = a
# df2[['LenCup'] < 10000] =1
c =df2[df2['LenCup']< 10000 ]
df2[df2['LenCup'] == df2['LenCup']< 10000] =1
b =df2[np.logical_and(df2['LenCup']> 10000,df2['LenCup']< 100000000)]
# df2
df2['LenCup'] = b['LenCup'] =2
df2
```

    - yugoslavia 국가의 월드컵 출전횟수가 2회인 선수들의 숫자를 구하시오.
    - 2002년도에 출전한 전체 선수 수를 구하시오.
    - 월드컵 출전 횟수가 1회 뿐인 선수들 중에서 가장 많은 득점을 올렸던 선수는 누구이며 몇 골을 넣었는지 구하시오.

```python
#yugoslavia 국가의 월드컵 출전횟수가 2회인 선수들의 숫자를 구하시오.
len(df2[np.logical_and(df2['Country'] == 'Yugoslavia',df2['LenCup'] >10000)])
```

```python
#- 2002년도에 출전한 전체 선수 수를 구하시오.
df2['Years'].str.replace('2002','출전')
df2[df2['Years'] == '2002'].head()
```

```python
#월드컵 출전 횟수가 1회 뿐인 선수들 중에서 가장 많은 득점을 올렸던 선수는 누구이며 몇 골을 넣었는지 구하시오.
df2[df2['LenCup']<10000][['Player','Goals']].sort_values(by=['Goals'],ascending=False).head(1)
```
