```python
df['구독자수'] = df['구독자수'].str.replace('만','0000',regex=False).astype('int64')
```

```python
df['조회수'] = df['조회수'].replace(['억','만'],['','0000'],regex = True).astype('int64')
df
```

```python
df['영상개수'] = df['영상개수'].replace([',','개'],['',''],regex = True).astype('int32')
```
