# 육상선수 종목별 기록 데이터

```python
import numpy as np

A = np.array([
              [3.73, 1.86, 15.8, 4.05, 7.27, 45.66, 34.92 ],
              [3.57, 1.80, 16.23, 2.96, 6.71, 42.56, 37.31],
              [3.22, 1.83, 14.20, 3.51, 6.68, 44.54, 39.23],
              [2.81, 1.80, 15.23, 2.69, 6.25, 42.78, 31.19],
              [2.91, 1.74, 14.76, 2.68, 6.32, 47.46, 35.53],
              [2.67, 1.83, 13.50, 1.96, 6.33, 42.82, 37.64]
])
```

# 허들 종목 기록 데이터 및 통계지표

```python
print('허들 종목 기록 데이터')
# 평균
print(f'평균:{A[:,0].mean():.3f}')
# 분산 var // ddof = 자유도 (n-ddof)
print(f'분산:{A[:,0].var(ddof=1):.3f}')
# 표준편차 std
print(f'표준편차:{A[:,0].std():.3f}')
```

# 공분산 행렬 구하기

```python
# rowvar : True(기본값)인 경우에는 행이 변수를 나타내며, 열은 각 변수의 관측치를 나타낸다
B = np.cov(A, rowvar=False)
print(B)
```

```python
col_mean = np.mean(A,axis= 0)
X = A - col_mean
cov = X.T.dot(X) / (len(X) - 1)
cov
```

# 선형대수
- 벡터 공간과 선형 변환을 중심으로 한 학문 체계를 말하며 다양한 분야에서 폭넓게 사용하고 있다.

## 벡터
- 데이터 여러개를 한 줄에 담아낼 수 있게 만든것
- 벡터에는 행벡터와 열벡터 두 종류가 있다.
- 서로 대응하는 성분끼리 덧셈과 뺄셈을 한다.
- 벡터의 모든 성분에 같은 수를 곱하는 것을 스칼라배라고 한다
- 서로 다른 차원의 벡터끼리는 덧셈과 뺄셈을 할 수 없다

## 유향선분
- 화살표 관점에서 벡터는 크기와 방향을 갖는 양이다
- 이러한 벡터에 대해 방향과 거리를 나타내는 화살표를 **유향선분**이라고 한다.

## 내적
- 벡터에서 서로 대응하는 성분끼리 곱한 다음, 그것들을 모두 더한 것이 내적이다.
- 벡터와 벡터의 내적은 수, 즉 스칼라가 된다.
- 서로 다른 차원의 벡터끼리는 계산을 할 수 없다.

- 정의
    - 내적을 정의할 대는 기하학적인 특징으로도 정의할 수도 있는대, 벡터 a와b가 이루는 각이O일 때 <a,b>는 다음과 같이 정의할 수 있다.
    - <a,b> = ||a||||b|| cos VEC

## 행렬의 곱셉
- 행렬 곱AB가 가능하기 위해서는 행렬 A의 열의 개수 와 행렬 B의 행의 개수가 같아야한다
- 교환법칙이 성립하지 않는다. AB != BA


## 선형변환
- 벡터 공간에서 또 다른 벡터 공간으로 벡터의 특징을 유지한 채 변환하는 방법
- 선이 선으로 남아 있어야 한다
- 원점이 고정된 채로 유지되어야 한다

## 고윳값과 고유 벡터
- 선형변환(일차변환)으로 방향이 바뀌지 않는 벡터를 고유벡터 라고 한다.
- 그리고, 이 고유벡터의 변환 전과 후의 비율을 고윳값이라고 한다
- 2X2 행렬은 반드시 두쌍의 고윳값과 고유벡터를 가진다
- 선형변환을 했을때 방향이 바뀌지 않고 크기(값)만 커질대 벡터의 방향이 고유벡터이고, 변화한 벡터의 길이 비율이 고윳값이다



## 고유값 분해
- 정방행렬의 인수분해를 고유값 분해라고 한다.
- 행렬A는 다음과 같이 고유값 분해 될 수 있다.
- 

- 입력데이터의 공분산 행렬을 구한다 X.T.X/(n-1)
- 공분산 행렬을 고유값 분해해서 고유 벡터와 고유값을 구한다 A = VAV**(1/2)
