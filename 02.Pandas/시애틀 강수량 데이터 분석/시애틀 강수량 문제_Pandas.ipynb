## 데이터 로딩

```python
import numpy as np
import pandas as pd

df = pd.read_csv('Seattle2014.csv')
df
```

```python
print(df.shape)
df.head()
```

```python
df.info()
```

## - 필요하지 않은 컬럼 삭제 (STATION, STATION_NAME)


```python
df =df.drop(columns = ['STATION','STATION_NAME'])
df
```

## DATE를 날짜 형식으로 수정 : 기존 DATE 컬럼의 타입 int -> str로 변경 후 날짜 형식으로 변경


```python
df['DATE'] =pd.to_datetime(df['DATE'].astype('str'))
df
```

## - 날짜 컬럼을 인덱스로 변경
  

```python
df.set_index('DATE',inplace = True)
```

## - 강수량('PRCP')이 많은 날 상위 10일 출력

```python
#df['PRCP'].sort_values()[::-1].head(10)
df.sort_values(by = 'PRCP',ascending = True)[['PRCP']].head(10)
```

## - 1월 강수량만 출력

```python
jan_df = df['PRCP'].iloc[:31]
jan_df
df['PRCP'][df.index.month == 1]
```

## - 1년동안 강수량이 40미만으로 내린 날이 몇일인가?

```python
alaks = df['PRCP'] < 40
df[alaks]['PRCP'].count()
len(df[df['PRCP']<40])
```

## - 1월에 내린 강수량의 합과 평균 강수량

```python
jan_df =df[:31]
jan_df[['PRCP']].agg(['sum','mean']).round(2)
#df[['PRCP']][df.index.month == 1].agg(['sum','mean']).round(2)
```
