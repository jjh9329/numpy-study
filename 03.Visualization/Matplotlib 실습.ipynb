## 기본적인 그래프 그리기

- plot() 함수의 인자로 표시할 데이터를 전달한 후 **show()** 함수를 통해 출력한다
- plot() 함수에 전달되는 리스트는 자동으로 y축의 값으로 설정된다.
- x축의 값은 0으로 시작해서 1의 간격으로 자동으로 설정된다.

## 스타일

```python
import numpy as np

plt.plot(np.arange(6),np.arange(6)**2,
        color ='red',
        linestyle = 'dashed',
        linewidth = 2.5,
        marker = 'o',
        markeredgecolor = 'green',
        markerfacecolor = 'yellow',
        markersize = 9.3)
plt.show()
```

# 서브플롯 그리기

- plt.subplot()함수를 이용해서 하나의 figure에 여러개의 그래프를 그릴 수있다.
- 주요 인자
    - nrow,ncol,index 를 지정하는 3가지 정수를 이용하여 서브플롯의 위치를 설정한다.

```python
import matplotlib.pyplot as plt
x1 = np.linspace(0,5)
x2 = np.linspace(0,2)

y1 = np.cos(2*np.pi * x1) * np.exp(-x1)
y2 = np.cos(2*np.pi * x2)

plt.figure(figsize=(11,4))

plt.subplot(1,2,1) #nrows =2 ,ncols =1,index =1
plt.plot(x1,y1,'o-')
plt.title('1st Graph')
plt.xlabel('tims (s)')
plt.ylabel('Damped')

plt.subplot(1,2,2) #nrows =2 ,ncols =1,index =2
plt.plot(x2,y2,'.-')
plt.title('2nd Graph')
plt.xlabel('tims (s)')
plt.ylabel('UnDamped')

plt.tight_layout() # 서브플롯의 레이아웃을 자동으로 맞춰준다.
plt.show()
```

- plt.subplots() 함수를 이용해서 서브플롯을 그릴 수 잇다.
- subplots() 함수는 Figure와 Axes 객체를 반환한다.
    - Axes 클래스 : 공간이 확보된 이미지 영역 정보를 가지고 있다.
- 주요 인자
    - nrows, ncols: 정수 값으로 서브플롯의 행,열수 지정
    - sharex, sharey: boolean 값으로(default: False) x,y축 사이의 공유 설정

```python
fig, ax = plt.subplots(1,4,sharey=True)

fig.set_size_inches((16,4)) # figure size 설정

ax[0].scatter(np.random.rand(100), np.random.rand(100), c ='cyan',edgecolor ='red', alpha =0.5) #산점도
ax[1].scatter(np.random.rand(100), np.random.rand(100), c ='magenta',edgecolor ='green', alpha =0.5) #산점도
ax[2].scatter(np.random.rand(100), np.random.rand(100), c ='yellow',edgecolor ='blue', alpha =0.5) #산점도
ax[3].scatter(np.random.rand(100), np.random.rand(100), c ='white',edgecolor ='black', alpha =0.5) #산점도

ax[0].set_title('subplot 0')
ax[1].set_title('subplot 1')
ax[2].set_title('subplot 2')
ax[3].set_title('subplot 3')
```

# 막대 그래프(bar chart) 그리기
- plt.bar()함수를 이용해서 막대 그래프를 그릴 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.arange(0,20)
y = x*2

plt.bar(x,y, label = 'Blue Bar', color = 'blue', edgecolor = 'yellow',linewidth =3)
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Chart Example')
x_ticks = np.arange(len(x))
plt.xticks(x_ticks,x_ticks+1) #(축 인덱스, 새로운 축 번호 값)
plt.legend()
plt.show()
```

# 히스토그램 그리기
- plt.hist() 함수를 이용해서 히스토그램을 그릴 수 있다.
- 자료의 분포를 몇 개의 구간으로 나눈 후 각 구간에 속하는 데이터의 빈도수를 표로 나타낸 것을 도수분포표(frequency table)라고한다

```python
y = np.random.randn(1000) #평균이 0, ㅍ준편차가 1인 표준 정규분포로부터 난수 1000개 생성

x = np.arange(len(y))
plt.bar(x,y)
plt.title('Raw Data <Bar Chart>')
plt.show()

plt.hist(y,bins = 50) # 표현할 데이터를 50개의 구간으로 나눈다.
plt.title('Histogram')
plt.show()

plt.hist(y,bins = 20, cumulative=True)
plt.title('Cumulative Histogram')
plt.show()
```

# 원 그래프(pie chart)그리기

- pie() 함수에 전달하는 값은 굳이 비율값을 전달하지 않고 일반적인 숫자 값을 전달하더라도 값들의 총 합에 대한 비율을 자동으로 계산해서 표시한다.

```python
labels = ['Python','Java','C']
# atuopct: wedge 위에 표시할 레이블 형식 (default = None)
plt.pie([0.3,0.35,0.4],labels=labels,autopct = '%.1f%%')
plt.title('Popular Language ranking')
plt.show()
```

```python
setion = [0.3,0.35,0.4]
labels = ['Python','Java','C']
color = ['coral','dodgerblue','yellowgreen']

plt.pie(setion, # 조각의 크기를 설정하는 값
        explode=(0.1,0,0), # 조각을 중심으로부터 얼마만큼 떨어지게 할것인지 설정(반지름의 10%만큼 벗어남을 의미)
        labels = labels, # 조각의 이름을 표시
        colors = color, # 조각의 색깔을 결정
        autopct = '%.2f%%', # 비율을 표시하는 형식 문자열을 지정
        shadow = True, # 그림자를 표시
        startangle=90, # 원형의 시작 부분을 각도로 표시.(default = 0,3시방향)
        counterclock= False, # 반시계 방향으로 순서대로 표시할지 결정 기본값은 True
        rotatelabels= True #조각의 이름을 회전시킨다(기본값은 False)
        )
plt.title('Pie Chart Example2')
plt.show()
```

# 박스 플롯 그리기 (상자 수염)
- plt.boxplot() 함수는 많은 데이터에서 데이터의 범위와 중앙값 등 통계 수치 값을 확인하고자 하는 목적으로 사용된다.
- 이상치(Outlier)가 있는지 검수하기 수월하다.
    - 최소값: 제 1사분위에서 1.5IQR(Inter Quantile Range: 사분위수 범위. Q3-Q1)을 뺀위치 
    - 제 1사분위(Q1) : 중앙값 기준으로 하위50% 중의 중앙값, 전체 데이터 중 하위 25%에 해당하는 값
    - 제 2사분위(Q2) : 50%의 위치로 중앙값(median)을 의미. 데이터의 정 가운데 순위에 해당하는 값
    - 제 3사분위(Q3) : 중앙값 기준으로 상위 50%중의 중앙값. 전체 데이터의 상위 25%에 해당하는 값
    - 최대값 : 제 3사분위에서 1.5QIR을 더한 위치
    - IQR(사분위 범위) : Q3 - Q1
    - 최대값, 최소값 범위를 벗어나는 값을 이상치 데이터로 간주한다.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data_a = np.random.normal(0,2.0,1000) # 평균 :0, 표준편차 :2 , 1000개
data_b = np.random.normal(-3,1.5,500) # 평균 :-3, 표준편차 :1.5 , 500개
data_c = np.random.normal(1.2,1.5,1500) # 평균 :1.2, 표준편차 :1.5 , 1500개

plt.boxplot([data_a,data_b,data_c])
plt.ylim(-10,10)
plt.xlabel('Data Type')
plt.ylabel('Value')
plt.show()
```

```python
np.random.seed(0)
data_a = np.random.normal(0,2.0,1000) # 평균 :0, 표준편차 :2 , 1000개
print(f'np.min():{np.min(data_a):.3f}, np.median():{np.median(data_a):.3f}, np.max():{np.max(data_a):.3f}')
quantile = np.quantile(data_a, [0,0.25,0.5,0.75,1])
print(quantile)
IQR = quantile[3] - quantile[1]
print('IQR: ',IQR)
print('MAX: ',quantile[3] + (1.5*IQR))
print('MIN: ',quantile[1] - (1.5*IQR))
```

```python
np.random.seed(0)

spread = np.random.rand(50)*100 # [0,100) 사이의 값 50개
center = np.ones(25) * 50 # 50(중앙값) 25개
filer_high = np.random.rand(10) * 100 + 100 # [100,200) 사이의 값 10개
filer_low = np.random.rand(10) * 100 - 100 # [-100,0) 사이의 값 10개

data =  np.concatenate([spread, center, filer_high, filer_low])
print(data.shape)

plt.boxplot(data)
plt.title("Box Plot - Outlier Check")
plt.show()
```
